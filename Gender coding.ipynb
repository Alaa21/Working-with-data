### Group1 (Janet1 - Daniela2)
Number of posts in common: All

```python
import datacompy
compare = datacompy.Compare(
    df1_1,
    df1_2,
    join_columns='SAMPLED CONTENT (URL to content)', 
)
print(compare.report())
```
